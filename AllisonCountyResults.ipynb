```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv("./project_voting/data/voting_VA.csv") #create variable with all voting data
data_df.head(20) #display first 20 rows
```

# **Create Df with wins by year**

```python
data_df['year'].value_counts() #check number of rows for each year
```

```python
#separate dataframe into dataframes by year
data_2000 = data_df[(data_df['year'] == 2000)]
data_2004 = data_df[(data_df['year'] == 2004)]
data_2008 = data_df[(data_df['year'] == 2008)]
data_2012 = data_df[(data_df['year'] == 2012)]
data_2016 = data_df[(data_df['year'] == 2016)]
data_2020 = data_df[(data_df['year'] == 2020)]
```

```python
data_by_year = [data_2000, data_2004, data_2008, data_2012, data_2016, data_2020] #list of each dataset
df_wins = pd.DataFrame() #create dataframe that will just include the winning rows (rows with highest votes for each year and county)
```

```python
#use a for loop to run through each dataframe per year
for year in data_by_year:
  wins = pd.DataFrame() #instantiate dataframe for the winning rows
  county_names = year['county_name'].values #create a list of the counties counted for each year
  county_names = list(set(county_names)) #ensure that list is unique (no multiples)
  for county in county_names: #loop through each county in each list
    county_rows = year[year['county_name'] == county] #pull out the rows that account for each county
    max_votes = county_rows['candidatevotes'].idxmax() #find the highest number of candidatevotes per row
    winning_row = year.loc[max_votes] #pull out the row with the highest number of candidate votes
    wins = pd.concat([wins, pd.DataFrame([winning_row])], ignore_index=True) #add winning row to wins dataframe
  df_wins = pd.concat([df_wins, wins], ignore_index = True) #add each winning row to large wins dataframe
```

# **Create Df with Dem and Rep votes per year**

```python
rep_dem_df = data_df[(data_df['party'] == 'DEMOCRAT') | (data_df['party'] == 'REPUBLICAN')] #create new dataframe with just the republican and democrat rows
```

```python
rep_dem_2020 = rep_dem_df[rep_dem_df['year'] == 2020] #pull out year 2020 since votes were counted differently (separated into different modes)
```

```python
county_names = rep_dem_2020['county_name'].values #create list of the counties counted in 2020
county_names = list(set(county_names)) #ensure that list is unqiue
#loop through each county in 2020 and sum up the votes for democrats and republicans so that the "ABSENTEE" mode includes the total number of votews
for county in county_names:
  dem_rows_sum = rep_dem_df[(rep_dem_df['county_name'] == county) & (rep_dem_df['year'] == 2020) & (rep_dem_df['party'] == 'DEMOCRAT')]['candidatevotes'].sum()
  rep_dem_df.loc[(rep_dem_df['county_name'] == county) & (rep_dem_df['year'] == 2020) & (rep_dem_df['party'] == 'DEMOCRAT') & (rep_dem_df['mode'] == 'ABSENTEE'), 'candidatevotes'] = dem_rows_sum
  rep_rows_sum = rep_dem_df[(rep_dem_df['county_name'] == county) & (rep_dem_df['year'] == 2020) & (rep_dem_df['party'] == 'REPUBLICAN')]['candidatevotes'].sum()
  rep_dem_df.loc[(rep_dem_df['county_name'] == county) & (rep_dem_df['year'] == 2020) & (rep_dem_df['party'] == 'REPUBLICAN') & (rep_dem_df['mode'] == 'ABSENTEE'), 'candidatevotes'] = rep_rows_sum
```

```python
rep_dem_df = rep_dem_df[(rep_dem_df['mode'] == 'TOTAL') | (rep_dem_df['mode'] == 'ABSENTEE')] #drop the rows that are not the "ABSENTEE" row
```

```python
rep_dem_df.loc[(rep_dem_df['mode'] != 'TOTAL'), 'mode'] = 'TOTAL' #rename "ABSENTEE" to match the other years which say "TOTAL"
```
