```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df_results = pd.read_csv('./project_voting/data/county_results.csv', low_memory = False)
```

```python
df_results
```

```python
years = [2000, 2004, 2008, 2012, 2016, 2020]
results_by_year = {}
for year in years:
    results_by_year[year] = df_results[df_results['year'] == year]
```

```python
party_results_by_year = {}
for year in years:
  party_results_by_year[year] = results_by_year[year]['party'].value_counts()
```
