```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv("./project_voting/data/voting_VA.csv")
data_df.head()
```

```python
data_df['year'].value_counts()
```

```python
data_2000 = data_df[(data_df['year'] == 2000)]
data_2004 = data_df[(data_df['year'] == 2004)]
data_2008 = data_df[(data_df['year'] == 2008)]
data_2012 = data_df[(data_df['year'] == 2012)]
data_2016 = data_df[(data_df['year'] == 2016)]
data_2020 = data_df[(data_df['year'] == 2020)]
```

```python
data_2000.head(2)
```

```python
county_names = data_2000['county_name'].values
```

```python
data_2000_ACC = data_2000[(data_2000['county_name'] == 'ACCOMACK')]
```

```python
data_2000_ACC.head(10)
```

```python
data_2000_wins = pd.DataFrame(columns=data_2000.columns)
```

```python
for county in county_names:
  county_row = data_2000[data_2000['county_name'] == county]
  max_votes = county_row['candidatevotes'].idxmax()
  winning_row = data_2000.loc[max_votes]
  data_2000_wins = data_2000_wins.append(winning_row, ignore_index = True)
```
